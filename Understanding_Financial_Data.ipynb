```
import pandas as pd
```

```
df = pd.read_csv('all_stocks_5yr.csv')
df.head()

ibm = df[df['Name']=='IBM']
ibm['close'].plot()
```

```
date = pd.date_range(df['date'].min(), df['date'].max())

close_prices = pd.DataFrame(index=date)

symbols = df['Name'].unique()

for symbol in symbols: 

  sym_data = df[df['Name']==symbol]
  aux_df = pd.DataFrame(data=sym_data['close'].to_numpy(),index=sym_data['date'],columns=[symbol])

  close_prices = close_prices.join(aux_df)

close_prices.head()
```

```
close_prices.to_csv('sp500_close.csv')
```
